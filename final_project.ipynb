```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import os
```

```python
df = pd.read_csv("2019.csv")

df.head()
```

```python
df.info()
```

```python
df.describe()
```

```python
top10 = df.sort_values(by="Score", ascending=False).head(10)

plt.figure()

sns.barplot(
    x="Score",
    y="Country or region",
    data=top10
)

plt.title("Top 10 Happiest Countries")
plt.show()
```

```python
plt.figure()

sns.scatterplot(
    x="GDP per capita",
    y="Score",
    data=df
)

plt.title("GDP per Capita vs Happiness Score")
plt.show()
```

```python
plt.figure()

sns.scatterplot(
    x="Healthy life expectancy",
    y="Score",
    data=df
)

plt.title("Life Expectancy vs Happiness Score")
plt.show()
```

```python
plt.figure()

corr = df.corr(numeric_only=True)

sns.heatmap(
    corr,
    annot=True
)

plt.title("Correlation Heatmap")
plt.show()
```

```python
least10 = df.sort_values(by="Score").head(10)

plt.figure()

sns.barplot(
    x="Score",
    y="Country or region",
    data=least10
)

plt.title("Least Happy Countries")
plt.show()
```
